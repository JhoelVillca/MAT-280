#4 Countplot Básico (Gender)

```python
sns.countplot(x = "Gender", data = mart)
```

#5 Countplot (Product line)

```python
plt.figure(figsize= (15, 5))
sns.countplot(x = "Product line", data = mart)
```

#6 Countplot Horizontal (Product line)

```python
sns.countplot(y = "Product line", data = mart)
```

#7 Countplot con Hue

```python
plt.figure(figsize= (15, 5))
sns.countplot(x = "Product line",hue = "Gender",  data = mart)
```

#8 Countplot con Paleta Personalizada

```python
plt.figure(figsize= (15, 5))
sns.countplot(x = "Product line",hue = "Gender",  data = mart, palette = 'inferno')
```

#9 Countplot con Estilo Outline

```python
plt.figure(figsize= (15, 5))
sns.countplot(x = "Product line", data = mart,
              facecolor = (0,0,0,0),
              linewidth = 3,
              edgecolor = sns.color_palette('inferno', 3)
              )
```
