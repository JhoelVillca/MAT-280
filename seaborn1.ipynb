```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#4 Countplot BÃ¡sico (Gender)

```python
sns.countplot(x = "Gender", data = mart)
```

#5 Countplot (Product line)

```python
plt.figure(figsize= (15, 6))
sns.countplot(x = "Product line", data = mart)
```

#6 Countplot Horizontal (Product line)

```python
sns.countplot(y = "Product line", data = mart)
```

#7 Countplot con Hue

```python
plt.figure(figsize= (15, 6))
sns.countplot(x = "Product line", hue = "Gender",data = mart)
```

#8 Countplot con Paleta Personalizada

```python
plt.figure(figsize= (15, 6))
sns.countplot(x = "Product line", hue = "Gender",data = mart, palette = 'cividis')
```

```python
plt.colormaps()
```

#9 Countplot con Estilo Outline

```python
plt.figure(figsize= (15, 6))
sns.countplot(x = "Product line", data = mart,
              facecolor = (0,0,0,0),
              linewidth = 3,
              edgecolor = sns.color_palette('cividis', 3)
              )
```
