```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris)
plt.show()
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris, s = 100)
plt.show()
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris, s = 100,
                  marginal_ticks=True, marginal_kws=dict(bins=25, fill = False)
                  )
g.fig.set_size_inches(8, 8)
g.set_axis_labels('largo del sepalo',  'ancho del sepalo', fontsize = 14)
plt.show()
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris, kind = "reg",
                  marginal_ticks=True, marginal_kws=dict(bins=25, fill = False)
                  )
g.fig.set_size_inches(8, 8)
g.set_axis_labels('largo del sepalo',  'ancho del sepalo', fontsize = 14)
plt.show()
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris, hue = "species",
                  marginal_ticks=True
                  )
g.fig.set_size_inches(8, 8)
g.set_axis_labels('largo del sepalo',  'ancho del sepalo', fontsize = 14)
plt.show()
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris, hue = "species", kind = "kde",
                  marginal_ticks=True
                  )
g.fig.set_size_inches(8, 8)
g.set_axis_labels('largo del sepalo',  'ancho del sepalo', fontsize = 14)
plt.show()
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris,  kind = "hex",
                  marginal_ticks=True
                  )
g.fig.set_size_inches(8, 8)
g.set_axis_labels('largo del sepalo',  'ancho del sepalo', fontsize = 14)
plt.show()
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris,  kind = "kde",
                  marginal_ticks=True
                  )
g.fig.set_size_inches(8, 8)
g.set_axis_labels('largo del sepalo',  'ancho del sepalo', fontsize = 14)
plt.show()
```

```python
sns.set_style("dark")
g = sns.jointplot(x = "sepal_length", y = "sepal_width", data = iris,  s = 100,
                  marginal_ticks=True
                  )
g.plot_joint(sns.kdeplot, color = "gray")
g.fig.set_size_inches(8, 8)
g.set_axis_labels('largo del sepalo',  'ancho del sepalo', fontsize = 14)
plt.show()
```
