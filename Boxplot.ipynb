```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mart = pd.read_excel("/content/supermarket_sales.xlsx")
mart.head()
```

```python
sns.boxplot(y = "Total", data = mart, width = 0.2)
```

```python
sns.boxplot(x = "Payment", y = "Total", data = mart)
```

```python
sns.boxplot(x = "Payment", y = "Total",hue = "Gender",  data = mart)
```

```python
sns.boxplot(x = "Payment", y = "Total",hue = "Gender",  data = mart, showmeans = True)
```

```python
sns.boxplot(x = "Payment", y = "Total",hue = "Gender",  data = mart, showmeans = True,
            meanprops = {"marker":"o", "markerfacecolor":"white", "markersize": "10", "markeredgecolor":"black"})
```

```python
sns.boxplot(x = "Payment", y = "Total",hue = "Gender",  data = mart, showmeans = True,
            meanprops = {"marker":"o", "markerfacecolor":"white", "markersize": "10", "markeredgecolor":"black"},
            palette= "CMRmap", linewidth= 5)
```

```python
plt.figure(figsize = (15,5))
sns.boxplot(data= mart)
```
