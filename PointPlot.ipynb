```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style = "whitegrid")
```

```python
ti = sns.load_dataset("titanic")
ti
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti)
```

```python
from numpy import median
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti, estimator = median)
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti, ci = 68)
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti, ci = 90)
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti, ci = "sd")
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti, hue = "sex")
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti, hue = "sex", palette= ["r","g"],  markers = ["o", "x"], linestyles=["-", "--"])
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti, hue = "sex", palette= ["r","g"],  markers = ["o", "x"], linestyles=["-", "--"], order = ["Third", "First", "Second"])
```

```python
plt.figure(figsize = (8, 6))
sns.pointplot(x = "class", y = "age", data = ti, capsize =  0.05, join = False)
```

```python
plt.figure(figsize = (8, 6))
sns.stripplot(x = "class", y = "age", data = ti)
sns.pointplot(x = "class", y = "age", data = ti, join = False, color = "red")
```

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x = "class", y = "age", data = ti)
sns.pointplot(x = "class", y = "age", data = ti, join = False, color = "red")
```

```python
plt.figure(figsize = (8, 6))
sns.violinplot(x = "class", y = "age", data = ti)
sns.pointplot(x = "class", y = "age", data = ti, join = False, color = "red")
```

```python
sns.catplot(x = "class", y = "age", data = ti,
            kind = "point", col ="embark_town")
```

```python
sns.catplot(x = "class", y = "age", data = ti,hue = "alive",
            kind = "point", col ="embark_town", height = 5, aspect=0.8)
```

```python
sns.catplot(x = "class", y = "age", data = ti,hue = "sex",
            kind = "point", col ="embark_town", height = 5, aspect=0.8)
```
